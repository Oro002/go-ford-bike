### Dataset Overview
> Bay Wheels is a regional public bicycle sharing system in California's San Francisco Bay Area. It is operated by Motivate in a partnership with the Metropolitan Transportation Commission and the Bay Area Air Quality Management District. Bay Wheels is 'the first regional and large-scale bicycle sharing system deployed in California and on the West Coast of the United States. It was established as Bay Area Bike Share in August 2013. As of January 2018, the Bay Wheels system had over 2,600 bicycles in 262 stations across San Francisco, East Bay and San Jose.

> In June 2017 the system was officially re-launched as Ford GoBike in a partnership with Ford Motor Company. After Motivate's acquisition by Lyft, the system was renamed to Bay Wheels in June 2019. The system is expected to expand to 7,000 bicycles around 540 stations in San Francisco, Oakland, Berkeley, Emeryville, and San Jose.*Source - [wikipedia](https://en.wikipedia.org/wiki/Bay_Wheels)*
This data set includes information about individual rides made in a bike-sharing system. 16 features were observed and 183412 entries were made.

```python
%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe
ford = codes.pd.read_csv('ford_bike_clean.csv')
```

```python
day_order = ford.start_day_of_the_week.value_counts().index
age_category_order= ford.age_category.value_counts().index
user_type_ordered = ford.member_gender.value_counts().index
```

## (Visualization 1)

> Thursday is the day with the most bike rides, bike rides are not often on weekends


```python
codes.plot_count(ford, x_axis = 'start_day_of_the_week', plot_order = day_order, plot_title ='count of rides by day of the week', 
          x_label = 'Day of the week', y_label = 'count', plot_color = green)
```

## (Visualization 2)

> Younger people between the age of 20 and 40 use the bike sharing service, there was no single record of any person less than 20 years old.

```python
codes.plot_count(ford, x_axis ='age_category', plot_order = age_category_order, show_count = True, y_label = 'count',
           x_label = 'Age group (child = (0 - 12), teen= (13 - 19), youth = (20 - 40), adult = (40 - 60), senior =(60 above))',
           plot_title = 'count of bike rides by age group', plot_color = green)
```

## (Visualization 3)
> It seems like bike riding and sharing is for the boys. Ha ha!

```python
codes.plot_count(ford, x_axis = 'member_gender', plot_color = green, show_count = True, plot_title = 'member gender count',
           x_label = 'member gender', y_label = " count", plot_order = user_type_ordered)
```

## (Visualization 4)
>customer travel longer hours while subscribers use the service more

```python
codes.sb_scatter_plot(ford, x_axis = 'duration_sec', y_axis = 'age', plot_hue = 'user_type', alpha = 1, marker ='D', 
                plot_title ='relationship between age, user type and bike ride duration', x_label ='Duration', y_label ='age',
               plot_figsize =(15,5))
```

## (Visualization 5)
>The boys are everywhere! Hold on! The ladies try to keep up too but it seems bike riding and sharing is mostly male dominated especially among the old.

```python
codes.sb_scatter_plot(ford, x_axis = 'duration_sec', y_axis = 'age', plot_hue = 'member_gender', alpha = 1, marker ='D', 
                plot_title ='relationship between age, gender and bike ride duration', x_label ='Duration', y_label ='age',
               plot_figsize =(15,5))
```
